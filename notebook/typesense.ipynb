RAG application using Typesense and Groq AI

```python
from dotenv import load_dotenv
import os
import pandas as pd
import typesense
```

```python
data_csv = pd.read_csv("../data/football_data/ginf.csv",encoding='UTF-8')
data_csv.drop(columns = ['odd_over','odd_under','odd_bts','odd_bts_n'], axis=1, inplace=True)
data_csv.to_json("../data/football_data/football_data.json", orient="records")
```

```python
data_dict = data_csv.to_dict(orient="records")
data_dict[0]
```

```python
load_dotenv()

client = typesense.Client({
    'nodes': [{
        'host': os.getenv('TYPESENSE_HOST'),
        'port': '443',
        'protocol': 'https'
    }],
    'api_key': os.getenv('TYPESENSE_API_KEY'),
    'connection_timeout_seconds': 2
})
```

```python
# book schema is used to create schema for the document collection

book_sachema = {
    'name': 'football_articles',
    'fields': [
        {'name': 'id_odsp', 'type': 'string'},
        {'name': 'link_odsp', 'type': 'string'},
        {'name': 'metadata', 'type': 'string', 'facet': True},
        {'name': 'adv_stats', 'type': 'bool', 'facet': True},
        {'name': 'date', 'type': 'string'},
        {'name': 'league', 'type': 'string'},
        {'name': 'season', 'type': 'int32'},
        {'name': 'country', 'type': 'string'},
        {'name': 'ht', 'type': 'string'},
        {'name': 'at', 'type': 'string'},
        {'name': 'fthg', 'type': 'int32'},
        {'name': 'ftag', 'type': 'int32'},
        {'name': 'odd_h', 'type': 'float'},
        {'name': 'odd_d', 'type': 'float'},
        {'name': 'odd_a', 'type': 'float'}
    ],
    'default_sorting_field': 'fthg'
}
print(client.collections.create(book_sachema))
```
